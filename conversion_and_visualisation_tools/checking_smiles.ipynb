```python
import pandas as pd

# File paths
csv_filename = r'smiles_only.csv'
txt_filename = r'hypotheticals_from_YOPXIL.txt'
# Read SMILES codes from the CSV file into a pandas DataFrame
df = pd.read_csv(csv_filename)
df
# Ensure the column name is 'SMILES' or adjust accordingly
if 'SMILES' not in df.columns:
    raise ValueError("The CSV file does not contain a 'SMILES' column.")

# Extract the SMILES column from the DataFrame as a set
csv_smiles_set = set(df['SMILES'])

# Read SMILES codes from the text file into a set
with open(txt_filename, 'r') as file:
    txt_smiles_set = set(line.strip() for line in file if line.strip())

# Find the intersection of both sets
common_smiles = csv_smiles_set.intersection(txt_smiles_set)

# Print out the common SMILES codes
print("SMILES codes present in both files:")
for smiles in common_smiles:
    print(f"Here: {smiles}")
```
